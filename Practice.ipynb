```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(11)
y = x ** 2
print(x, y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y, 'r--*')
plt.show()
```

```python
plt.plot(x,y, 'r--*')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(range(0,11))
plt.yticks(range(0,101,10))
plt.title('Figure/plot title')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, label="x**2")
axes.plot(x, x**3, label="x**3")
axes.legend(loc=2)
plt.show()
```
